```python
import os
import random
import cv2
import matplotlib.pyplot as plt

# move one level up from notebooks/
BASE_DIR = os.path.join("..", "data", "raw", "kaggle", "training")
BASE_DIR = os.path.abspath(BASE_DIR)

print("BASE_DIR resolved to:", BASE_DIR)

classes = ["glioma", "meningioma", "notumor", "pituitary"]

plt.figure(figsize=(10, 10))

for i, cls in enumerate(classes):
    cls_dir = os.path.join(BASE_DIR, cls)
    print("Reading from:", cls_dir)

    img_name = random.choice(os.listdir(cls_dir))
    img_path = os.path.join(cls_dir, img_name)

    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 2, i + 1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis("off")

plt.suptitle("Brain MRI Samples (Kaggle Dataset)", fontsize=16)
plt.show()
```
